```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Heart_Disease.csv")
df = df.drop(columns=['education'])
df
```

```python
df.isnull().sum()
```

```python
df['Heart Disease']=df['Heart Disease'].astype('category')
df['Heart Disease']=df['Heart Disease'].cat.codes

df['diabetes']=df['diabetes'].astype('category')
df['diabetes']=df['diabetes'].cat.codes

df['smokingStatus']=df['smokingStatus'].astype('category')
df['smokingStatus']=df['smokingStatus'].cat.codes

df['Gender']=df['Gender'].astype('category')
df['Gender']=df['Gender'].cat.codes
```

```python
df
```

```python
df['smokingStatus'] = df['smokingStatus'].replace(-1, np.nan)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Heart Disease', axis=1)
y = df['Heart Disease']
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test)
print(y_predicted)
print(y_test)
```

```python
print(y_predicted)
print(y_test.values)
```

```python
X = df[['age']].values
y = df['Heart Disease'].values
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
x_vals = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_pred = model.predict_proba(x_vals)[:, 1]
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['Heart Disease'], color='blue', alpha=0.5, label='Data points')
plt.plot(x_vals, y_pred, color='red', linewidth=2, label='Logistic Regression Curve')
plt.title("Logistic Regression: Age vs Heart Disease")
plt.xlabel("Age")
plt.ylabel("Probability of Heart Disease")
plt.legend()
plt.grid(True, alpha=0.6)
plt.show()
```
