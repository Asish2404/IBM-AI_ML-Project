```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Heart_Disease.csv")
df = df.drop(columns=['education'])
df
```

```python
df.isnull().sum()
```

```python
df['Heart Disease']=df['Heart Disease'].astype('category')
df['Heart Disease']=df['Heart Disease'].cat.codes

df['diabetes']=df['diabetes'].astype('category')
df['diabetes']=df['diabetes'].cat.codes

df['smokingStatus']=df['smokingStatus'].astype('category')
df['smokingStatus']=df['smokingStatus'].cat.codes

df['Gender']=df['Gender'].astype('category')
df['Gender']=df['Gender'].cat.codes
```

```python
df
```

```python
df['smokingStatus'] = df['smokingStatus'].replace(-1, np.nan)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Heart Disease', axis=1)
y = df['Heart Disease']
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test)
print(y_predicted)
print(y_test)
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8,5))
plt.hist(df['age'],color="blue", bins=20, alpha=0.6, label='Age Distribution')
plt.title("Age vs Heart Disease")
plt.xlabel("Age")
plt.ylabel("Count / Probability")
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

```python
df['BMI_bin'] = pd.cut(df['BMI'], bins=7)
bar_vals = df.groupby('BMI_bin', observed=True)['Heart Disease'].mean()
plt.figure(figsize=(8,5))
plt.bar(bar_vals.index.astype(str), bar_vals, alpha=0.6)
plt.title("Heart Disease Probability by BMI Range")
plt.xlabel("BMI Range")
plt.ylabel("Avg Probability")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.6)

plt.show()
```

```python
if 'Chol_bin' in df.columns:
    df = df.drop(columns=['Chol_bin'])
bins = [0, 150, 200, 250, 300, df['totChol'].max()]
labels = ["<150", "150-200", "200-250", "250-300", ">300"]
df.loc[:, 'Chol_bin'] = pd.cut(df['totChol'], bins=bins, labels=labels, include_lowest=True)
bar_vals = df.groupby('Chol_bin', observed=True)['Heart Disease'].mean()
plt.figure(figsize=(8,5))
plt.bar(labels, bar_vals)
plt.title("Heart Disease Probability by Total Cholesterol Range")
plt.xlabel("Total Cholesterol Range")
plt.ylabel("Avg Heart Disease Probability")
plt.show()
```

```python
df.loc[:, 'sysBP_bin'] = pd.cut(df['sysBP'], bins=7)
bar_vals = df.groupby('sysBP_bin', observed=True)['Heart Disease'].mean()
plt.figure(figsize=(8,5))
plt.bar(range(len(bar_vals)), bar_vals, color='#2596be', alpha=0.7)
plt.xticks(range(len(bar_vals)), bar_vals.index.astype(str), rotation=45)
plt.title("Systolic BP vs Heart Disease")
plt.xlabel("Systolic BP Range")
plt.ylabel("Avg Probability of Heart Disease")
plt.grid(True, alpha=0.6)
plt.show()
```

```python
if 'cigs_bin' in df.columns:
    df = df.drop(columns=['cigs_bin'])
bins = [0, 1, 5, 10, 20, df['cigsPerDay'].max()]
labels = ["0", "1-5", "6-10", "11-20", "more than 20"]
df['cigs_bin'] = pd.cut(df['cigsPerDay'], bins=bins, labels=labels, include_lowest=True)
pie_vals = df.groupby('cigs_bin', observed=True)['Heart Disease'].mean()
plt.figure(figsize=(7,7))
plt.pie(
    pie_vals,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"linewidth": 1, "edgecolor": "white"}
)
plt.title("Heart Disease Probability by Cigarettes per Day")
plt.show()
```
